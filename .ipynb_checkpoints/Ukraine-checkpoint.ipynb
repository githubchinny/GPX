From GeoPandas natural earth default maps get the shape of Ukraine, which doesn't include Crimea

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
Ukraine = world[world.name == "Ukraine"]

base = world.plot(figsize=(20,10),color='white', edgecolor='black')
Ukraine.plot(ax=base)
```

I can then find the point data in gdf_places that are outside of the Ukraine shape which doesn't include Crimea peninsula

```python
my_union = Ukraine.geometry.unary_union
my_union
```

```python
gdf_places = gdf_places.loc[~gdf_places.geometry.apply(lambda x: my_union.contains(x))]
```

This leaves the following places for Crimea Peninsula, and a few stragglers around the perimeter of Ukraine.

```python
gdf_places.shape
```

Map each of the points onto a map as a cirle.  There is work to be done here, with labelling and popups that could be adding to provide more information.

```python
def add_markers(mapobj, gdf):
    coords = []
    for i, row in gdf.iterrows():
        coords.append([row.geometry.y, row.geometry.x])
    for coord in coords:
        folium.CircleMarker(location = coord,
                            radius = 2.5, 
                            fill = True,
                            fill_color = '#F50057',
                            fill_opacity = 0.75,
                            color = 'whitesmoke',
                            weight = 0.5).add_to(mapobj)
    return mapobj

f = folium.Figure(height = 400)
m = folium.Map([crimea_lat, crimea_long], zoom_start = 8, tiles='Cartodb dark_matter')
m.add_to(f)

add_markers(m, gdf_places)

embed_map(m, "Markers.html")
```

And this is a HeatMap or density map, indicating where the points are more densely located ... if you couldn't work that out from the map above!

```python
def add_markers(mapobj, gdf):
    heat_data = [[row.geometry.centroid.y,row.geometry.centroid.x] for index, row in gdf.iterrows()]
    HeatMap(data=heat_data, radius=16).add_to(mapobj)
    Marker([crimea_lat,crimea_long]).add_to(mapobj)
    
f = folium.Figure(height = 400)
m = folium.Map([crimea_lat, crimea_long], zoom_start = 6)
m.add_to(f)

add_markers(m, gdf_places)

embed_map(m, "HeatMap.html")
```
